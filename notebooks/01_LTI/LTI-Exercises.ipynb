```python
fsig = 50.0      # signal frequenc
tstep = 1e-3     # time step for "analog" signal
Tmax = 8.0/fsig  # timespan for 6 signal periods 

t = np.arange(0,Tmax,tstep)  # time t
```

```python
# xtemp(t) is continuous signal of 1.5V
xtemp = [1.5]*len(t) # signal x(t)
xn = 0.5*np.cos(2*np.pi*fsig*t)
xt = xtemp + xn  # signal x(t)
```

```python
fig, ax = plt.subplots(**size) 
ax.set_xlabel(r'Time (s) $\rightarrow$')
ax.set_ylabel(r'Amplitude (V) $\rightarrow$')
ax.plot(t, xt, 'b-');
ax.plot(t, xn, 'r-');
ax.plot(t, xtemp, 'r-');
```

```python
Ts1 = 1 / 25
Ts2 = 1 / 200
Ts3 = 1 / 240
Ts4 = 1 / (np.pi * 50)
```

```python
N_Ts = Tmax / Ts1
n = np.arange(0,round(N_Ts))

xn = 1.5 + 0.5 * np.cos(2*np.pi*fsig*n*Ts1)

fig, ax = plt.subplots(**size) 
ax.plot(t, xt, 'b-') # x(t) with blue line
ax.stem(n*Ts1, xn, linefmt='r-', markerfmt='ro', basefmt='k');
```
